```python
#Importing dependencies

import numpy as np
import tensorflow as tf
import pandas as pd
import tensorflow.keras as keras

!pip install gensim
#!pip install google.cloud
#import gzip
import gensim.downloader as api

# Load pre-trained GloVe vectors (300-dimensional)
model = api.load('glove-wiki-gigaword-300')

result = model.most_similar(positive=['japan','paris'], negative=['france'],topn=1)
print("Japan - France + Paris = ",result[0][0])
print("similarity=",result[0][1])
result = model.most_similar(positive=['king', 'woman'], negative=['man'],topn=1)
print("King - Man + Woman = ", result[0][0])
print("similarity=",result[0][1])

#Check Similarity
result = model.most_similar(positive=['programming'],topn=5)
print("top five words similar to 'programming'")
for word,similarity in result:
  print(word,similarity)
```
